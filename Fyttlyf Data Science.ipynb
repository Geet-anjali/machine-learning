```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
df_count = df[df["Event_Key"]==1].groupby(["Group"]).sum()
df_count
```

```python

from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
y = df_count["Event_Key"]
y
```

```python
ARMAmodel = SARIMAX(y, order = (1, 0, 1))
```

```python
ARMAmodel = ARMAmodel.fit()
```

```python
y_pred = ARMAmodel.get_forecast(3)
y_pred_df =pd.DataFrame(y_pred.conf_int(alpha = 0.05) )
y_pred_df
```

```python
len(X_test.index)
```

```python
y_pred_df["Predictions"] = ARMAmodel.predict(start = y_pred_df.index[0], end = y_pred_df.index[-1])
y_pred_df.index = [4,5,6]
y_pred_out = y_pred_df["Predictions"] 
```

```python
plt.plot(y_pred_out, color='green', label = 'Predictions')
plt.legend()
```

```python
SARIMAXmodel = SARIMAX(y, order = (5, 4, 2), seasonal_order=(2,2,2,12))
SARIMAXmodel = SARIMAXmodel.fit()

y_pred = SARIMAXmodel.get_forecast(3)
y_pred_df = pd.DataFrame(y_pred.conf_int(alpha = 0.05))
y_pred_df["Predictions"] = SARIMAXmodel.predict(start = y_pred_df.index[0], end = y_pred_df.index[2])
y_pred_df.index = [4,5,6]
y_pred_out = y_pred_df["Predictions"] 
plt.figure(figsize=[10,10])
palette = sns.color_palette("tab10", 4)
plt.plot(y_pred_out, color='Blue', label = 'SARIMA Predictions')
plt.xlabel("Group")
plt.ylabel("Count of clicks")


plt.legend()
plt.show()
```
