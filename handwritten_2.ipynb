# Task1

You have to find average of price1 on same week and create a csv file with two columns
date,average price.

```python
df['date'] = pd.to_datetime(df['date'])
#convert to weeks and for starting with 1 add factorize
df['week'] = pd.factorize(df['date'].dt.weekofyear)[0] + 1
print (df)
```

```python

df2=df.groupby('week')['price1'].sum()
df2
```

```python
df3=df.groupby('week')['week'].count()
df3
```

```python
df_average= df2/df3
df_average
```

```python
Average =[]
for i in range(df['date'].count()):
    Average.append(df_average[df.loc[i,'week']])
    
```

```python
data = {'Date':df['date'], 'Average':Average}
df_t1 = pd.DataFrame(data)
```

```python
df_t1
```

```python
df_t1.to_csv('task1.csv')
```
